```python
import pandas as pd

def leer_excel_nombres_profes_a_arreglo(ruta_excel, nombre_hoja, columnas):
    # Leer el archivo de Excel
    df = pd.read_excel(ruta_excel, sheet_name=nombre_hoja, usecols=columnas)

    # Procesar la columna "Nombres" para quedarse solo con el primer nombre
    df['NOMBRES'] = df['NOMBRES'].apply(lambda x: x.split()[0])

    # Procesar la columna "Nombres" para quedarse solo con el primer nombre
    df['PRIMER APELLIDO'] = df['PRIMER APELLIDO'].apply(lambda x: x.split()[0])

    # Convertir el DataFrame a una lista de listas
    arreglo = df.values.tolist()

    return arreglo

# Ejemplo de uso:
ruta_excel = 'imported_files/BBDD ACADÉMICOS FIN-PUCV.xlsx'
nombre_hoja = 'DATA 1S2024'
columnas = ['NOMBRES', 'PRIMER APELLIDO']  # Especificar las columnas que deseas copiar

arreglo_profesores_fin = leer_excel_nombres_profes_a_arreglo(ruta_excel, nombre_hoja, columnas)

# Imprimir el arreglo bidimensional
for fila in arreglo_profesores_fin:
    print(fila)
```

```python
def leer_excel_proyectos_anid_a_arreglo(ruta_excel, nombre_hoja, columnas, nombre_columna_a_filtrar, valores_filtro):
    # Leer el archivo de Excel
    df = pd.read_excel(ruta_excel, sheet_name=nombre_hoja, usecols=columnas)

    # Filtrar las filas donde el valor de la columna especificada sea igual a cualquiera de los valores de filtro
    df_filtrado = df[df[nombre_columna_a_filtrar].isin(valores_filtro)]

    # Convertir el DataFrame filtrado a una lista de listas
    arreglo_filtrado = df_filtrado.values.tolist()

    return arreglo_filtrado

# Ejemplo de uso:
# Nota: Modificar el nombre del archivo acorde al nombre del archivo .xlsx descargado de https://github.com/ANID-GITHUB/Historico-de-Proyectos-Adjudicados/tree/master
ruta_excel = 'imported_files/BDH_PROYECTOS_2003_2022.xlsx'
nombre_hoja = 'Sheet1'
columnas = ['CODIGO_PROYECTO', 'SUBDIRECCION', 'PROGRAMA', 'INSTRUMENTO', 'NOMBRE_CONCURSO', 'AGNO_CONCURSO', 'AGNO_FALLO', 'NOMBRE_PROYECTO', 'NOMBRE_RESPONSABLE', 'INSTITUCION_PRINCIPAL', 'MONTO_ADJUDICADO']  # Especificar las columnas que deseas copiar

nombre_columna_a_filtrar = 'INSTITUCION_PRINCIPAL'  # Nombre de la columna a filtrar
valores_filtro = ["PONTIFICIA UNIVERSIDAD CATÓLICA DE VALPARAÍSO", "PONTIFICIA UNIVERSIDAD CATOLICA DE VALPARAISO"]  # Valor a filtrar

arreglo_proyectos_PUCV_filtrado = leer_excel_proyectos_anid_a_arreglo(ruta_excel, nombre_hoja, columnas, nombre_columna_a_filtrar, valores_filtro)

# Imprimir el arreglo bidimensional
for fila in arreglo_proyectos_PUCV_filtrado:
    print(fila)
```

```python
def comparar_arreglos(arreglo_profesores_fin, arreglo_filtrado, indice_columna_responsable):
    resultados_comparacion = []
    for profesor in arreglo_profesores_fin:
        nombre_pila, apellido = profesor
        for fila in arreglo_filtrado:
            nombre_responsable = fila[indice_columna_responsable]
            if nombre_pila in nombre_responsable and apellido in nombre_responsable:
                resultados_comparacion.append(fila)
                break
    return resultados_comparacion

# El índice de la columna 'nombre_responsable' en el arreglo filtrado (suponiendo que es la columna 6)
indice_columna_responsable = 8

# Comparar los arreglos
resultados_comparacion = comparar_arreglos(arreglo_profesores_fin, arreglo_proyectos_PUCV_filtrado, indice_columna_responsable)

# Imprimir los resultados de la comparación
for fila in resultados_comparacion:
    print(fila)
```

```python
def exportar_a_excel(resultados, nombre_columnas, ruta_salida):
    # Crear un DataFrame con los resultados
    df_resultados = pd.DataFrame(resultados, columns=nombre_columnas)

    # Exportar el DataFrame a un archivo Excel
    df_resultados.to_excel(ruta_salida, index=False)

# Exportar los resultados a un archivo Excel
nombre_columnas = ['CODIGO_PROYECTO', 'SUBDIRECCION', 'PROGRAMA', 'INSTRUMENTO', 'NOMBRE_CONCURSO','AGNO_CONCURSO', 'AGNO_FALLO', 'NOMBRE_PROYECTO', 'NOMBRE_RESPONSABLE', 'INSTITUCION_PRINCIPAL', 'MONTO_ADJUDICADO']
ruta_salida = 'proyectos_ANID_adjudicados_por_profesores_FIN.xlsx'

exportar_a_excel(resultados_comparacion, nombre_columnas, ruta_salida)
```
